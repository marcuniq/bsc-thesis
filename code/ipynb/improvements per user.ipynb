```python
import pandas as pd
```

```python
mpcf_si_metrics = pd.read_csv('../metrics/mpcf-si/2016-06-30_07.36.22_si_ml-1m_e20_tt-0.7_train-4_user-metrics.csv')
mpcf_metrics = pd.read_csv('../metrics/mpcf/2016-06-22_19.59.20_no-si_ml-1m_e20_tt-0.7_train-4_task-0_user-metrics.csv')
```

```python
columns = mpcf_si_metrics.columns
for col in columns:
    mpcf_si_metrics.rename(columns={col: 'si_{}'.format(col)}, inplace=True)
```

```python
mpcf_si_metrics.columns
```

```python
merged_metrics = mpcf_metrics.merge(mpcf_si_metrics, left_on='user_id', right_on='si_user_id', how='inner')
```

```python
merged_metrics
```

```python
merged_metrics['auc_ratio'] = merged_metrics['si_auc'] / merged_metrics['auc']
```

```python
merged_metrics['auc_ratio'].max()
```

```python
merged_metrics['auc_ratio'].min()
```

### users for which AUC was improved by using side information

```python
mm = merged_metrics[merged_metrics['auc_ratio'] >= 1][['user_id','auc', 'si_auc', 'auc_ratio', 'nb_train_ratings', 'nb_test_ratings']]
mm.sort_values('auc_ratio', ascending=False)
```

```python
mm['nb_train_ratings'].mean()
```

```python
mm['auc'].mean()
```

```python
mm['si_auc'].mean()
```

### users for which AUC declined by using side information

```python
mm = merged_metrics[merged_metrics['auc_ratio'] < 1][['user_id','auc', 'si_auc', 'auc_ratio', 'nb_train_ratings', 'nb_test_ratings']]
mm.sort_values('auc_ratio', ascending=True)
```
